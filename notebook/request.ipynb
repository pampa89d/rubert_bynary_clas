```python
import requests
```

```python
url = "http://localhost:8000/spam_classification/"
data = {
    "messages": ["Срочно купи биткоин и выходи к нам на работу!", "Завтра будет совещание в 10:00"]
}

response = requests.post(url, json=data)

if response.status_code == 200:
    print("Результат:", response.json())
else:
    print(f"Ошибка {response.status_code}: {response.text}")
```
